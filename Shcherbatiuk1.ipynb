# **Построение графиков в размерных переменных**
**Щербатюк Р.Р.**

```python
import numpy as np
import scipy.special as sc
import matplotlib.pyplot as plt
```

```python
def p_ss_atma(p_res_atma = 260,
                 q_liq_sm3day = 40,
                 mu_cP = 1,
                 B_m3m3 = 1.2,
                 k_mD = 40,
                 h_m = 10,
                 r_e = 240,
                 r = 0.1):
  return p_res_atma - 18.41 * q_liq_sm3day*mu_cP*B_m3m3/k_mD/h_m * np.log(r_e/r)
```

```python
r = np.arange(0.1, 100, 1)

plt.plot(r, p_ss_atma(r=r))
plt.show()
```

**Воронка** **депрессии**

```python
def p_rt_atma(p_res_atma = 260,
                 q_liq_sm3day = 40,
                 mu_cP = 1,
                 B_m3m3 = 1.2,
                 k_mD = 40,
                 h_m = 10,
                 r_e = 240,
                 r = 0.1,
                 y = 0.15,
                 ct = 10**(-5),
                 t_h = 50):
  return p_res_atma - 18.41 * q_liq_sm3day*mu_cP*B_m3m3/k_mD/h_m * -1/2*sc.expi(-y*mu_cP*ct*r**2 / 0.00144/k_mD / t_h)
```

```python
r = np.arange(0.1, 100, 1)

plt.text(90, 242, '50 ч', fontsize=10)
plt.text(90, 238.75, "500 ч", fontsize=10)
plt.text(90, 235.5, "5000 ч", fontsize=10)
plt.plot(r, p_rt_atma(r=r, t_h=50))
plt.plot(r, p_rt_atma(r=r, t_h=500))
plt.plot(r, p_rt_atma(r=r, t_h=5000))
plt.title('Воронка депрессии', fontsize=14, fontweight="bold")
grid1 = plt.grid(True)
plt.xlabel("$\it{Расстояние, м}$")
plt.ylabel("$\it{Пластовое давление, атм}$")
plt.show()
```

**График зависимости забойного давления скважины от времени**

```python
def p_z_atma(p_res_atma = 260,
                 q_liq_sm3day = 40,
                 mu_cP = 1,
                 B_m3m3 = 1.2,
                 k_mD = 40,
                 h_m = 10,
                 r_e = 240,
                 r = 0.1,
                 y = 0.15,
                 ct = 10**(-5),
                 t_h = 50):
  return p_res_atma + 18.41 * q_liq_sm3day*mu_cP*B_m3m3/k_mD/h_m * 1/2*sc.expi(-y*mu_cP*ct*r**2 / 0.00144/k_mD / t_h)
```

```python
t = np.arange(1, 1e6, 1e4)

plt.plot(t, p_z_atma(t_h=t, k_mD=40), label='40 мД')
plt.plot(t, p_z_atma(t_h=t, k_mD=80), label='80 мД')
plt.plot(t, p_z_atma(t_h=t, k_mD=120), label='120 мД')
plt.title('Изменение забойного давления', fontsize=14, fontweight="bold")
grid1 = plt.grid(True)
plt.xlabel("$\it{Время, м}$")
plt.ylabel("$\it{Забойное давление, атм}$")
plt.legend()
plt.show()
```

```python
t = np.arange(1, 1e6, 1e4)

plt.plot(t, p_z_atma(t_h=t, k_mD=40), label='40 мД')
plt.plot(t, p_z_atma(t_h=t, k_mD=80), label='80 мД')
plt.plot(t, p_z_atma(t_h=t, k_mD=120), label='120 мД')
plt.title('Изменение забойного давления', fontsize=14, fontweight="bold")
grid1 = plt.grid(True)
plt.xlabel("$\it{log t}$")
plt.ylabel("$\it{Забойное давление, атм}$")
plt.xscale('log')
plt.legend()
plt.show()
```

```python
t = np.arange(1, 1e6, 1e4)

plt.plot(t, p_z_atma(t_h=t))
plt.title('Изменение депрессии во времени', fontsize=14, fontweight="bold")
grid1 = plt.grid(True)
plt.xlabel("$\it{log t}$")
plt.ylabel("$\it{log p}$")
plt.xscale('log')
plt.yscale('log')
plt.show()
```

**График логарифмической производной депрессии от времени в двойных логарифмических координатах**

```python
t = np.arange(1, 1e6, 1e4)
u = abs(np.diff(p_z_atma(t_h=t))*t[:-1])

plt.plot(t[:-1], u)

plt.title('Логарифмическая производная депрессии во времени', fontsize=14, fontweight="bold")
grid1 = plt.grid(True)
plt.xlabel("$\it{t}$")
plt.ylabel("$\it{p}$")

plt.xscale('log')
plt.yscale('log')






plt.show()





```
